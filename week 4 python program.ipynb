```python
#truple
print("----tuple example---")
coordinates=(10.5,20.5,10)
print("tuple:",coordinates)
print("first coordinate:",coordinates[0])
print(coordinates[-1])
x,y,z=coordinates
print(f"unpacked:x={x},y={y},z={z}")
```
