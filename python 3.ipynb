```python
""" Given an integer n: 
● If n is odd → print "Weird" 
● If n is even and in range 2–5 → "Not Weird" 
● If n is even and in range 6–20 → "Weird" 
● If n is even and > 20 → "Not Weird"
"""
n = int(input("Enter an integer: "))

if n % 2 != 0:
    print("Weird")
else:
    if 2 <= n <= 5:
        print("Not Weird")
    elif 6 <= n <= 20:
        print("Weird")
    elif n > 20:
        print("Not Weird")
```

```python
"""
Triangle Validity 
Given 3 sides a, b, c, check: 
● If triangle is possible (sum of any 2 > third) 
● Print "Valid" or "Invalid"
"""
a = int(input("Enter side a: "))
b = int(input("Enter side b: "))
c = int(input("Enter side c: "))

if (a + b > c) and (a + c > b) and (b + c > a):
    print("Valid")
else:
    print("Invalid")
```

```python
"""
Type of Triangle 
Given 3 sides: 
● Equilateral (all equal) 
● Isosceles (any two equal) 
● Scalene (all different)
"""
a = float(input("Side a: "))
b = float(input("Side b: "))
c = float(input("Side c: "))

if a == b == c:
    print("Equilateral")
elif a == b or b == c or a == c:
    print("Isosceles")
else:
    print("Scalene")
```

```python
"""
 Electricity Bill Calculation 
Input units consumed: 
● First 100 → ₹5/unit 
● Next 100 → ₹7/unit 
● Above 200 → ₹10/unit 
Calculate total bill.
"""
units = float(input("Enter units consumed: "))
bill = 0

if units <= 100:
    bill = units * 5
elif units <= 200:
    # First 100 at 5, remaining at 7
    bill = (100 * 5) + ((units - 100) * 7)
else:
    # First 100 at 5, next 100 at 7, remaining at 10
    bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)

print(f"Total Bill: ₹{bill}")
```

```python
"""
Salary Bonus Calculation 
Input salary and years of service: 
● ≥ 5 years → 10% bonus 
● < 5 years → 5% bonus 
Print final salary.
"""
salary = float(input("Enter salary: "))
years = int(input("Enter years of service: "))
if years > 5:
    bonus = 0.05 * salary
    print(f"Bonus amount: {bonus}")
    print(f"Total Salary: {salary + bonus}")
else:
    print("No bonus applicable.")
```

```python
"""
 ATM Withdrawal System 
Input: 
● Account balance 
● Withdrawal amount 
Conditions: 
● Amount must be multiple of 100 
● Balance must be sufficient 
● Minimum balance ₹500 should remain 
Print "Transaction Successful" or "Transaction Failed"
"""
balance = float(input("Enter account balance: "))
amount = int(input("Enter withdrawal amount: "))

if amount % 100 == 0 and (balance - amount) >= 500:
    print("Transaction Successful")
else:
    print("Transaction Failed")
```

```python
"""
Quadratic Equation Roots Nature 
Given a, b, c, calculate discriminant D = b² - 4ac: 
● D > 0 → "Real and Distinct" 
● D = 0 → "Real and Equal" 
● D < 0 → "Imaginary"
"""
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

D = (b**2) - (4*a*c)

if D > 0:
    print("Real and Distinct")
elif D == 0:
    print("Real and Equal")
else:
    print("Imaginary")
```

```python
"""
Discount Calculator 
Input purchase amount: 
● ≥ 5000 → 20% discount 
● 2000–4999 → 10% 
● < 2000 → no discount 
Print final amount. 
"""
amount = float(input("Enter purchase amount: "))

if amount >= 5000:
    discount = 0.20
elif amount >= 2000:
    discount = 0.10
else:
    discount = 0.0

final_amount = amount - (amount * discount)
print(f"Final amount to pay: {final_amount}")
```

```python
"""
 Menu Driven Program 
Display menu: 
1. Check even/odd 
2. Check positive/negative 
3. Exit 
Take user choice and perform operation using if-elif.
"""
while True:
    print("\n--- Menu ---")
    print("1. Check even/odd")
    print("2. Check positive/negative")
    print("3. Exit")
    
    choice = input("Enter your choice (1-3): ")

    if choice == '1':
        num = int(input("Enter a number: "))
        if num % 2 == 0:
            print(f"{num} is Even")
        else:
            print(f"{num} is Odd")
            
    elif choice == '2':
        num = float(input("Enter a number: "))
        if num > 0:
            print("Positive")
        elif num < 0:
            print("Negative")
        else:
            print("Zero")
            
    elif choice == '3':
        print("Exiting program...")
        break
        
    else:
        print("Invalid choice! Please try again.")
```
