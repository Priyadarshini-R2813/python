```python
"""Two Sum (Optimized) 
Problem: Given an array and a target
return indices of two numbers whose sum = target.
Constraint: Must be O(n) (no nested loops)"""
def two_sum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        diff = target - num
        if diff in seen:
            return [seen[diff], i]
        seen[num] = i
print(two_sum([2,7,11,15],9))
```

```python
"""
Longest Substring Without Repeating Characters Problem: 
Input a string → find length of longest substring with all unique characters
Example: "abcabcbb" → Output: 3"""
def length_of_longest_substring(s):
    seen = {}
    left = 0
    max_len = 0

    for right, ch in enumerate(s):
        if ch in seen and seen[ch] >= left:
            left = seen[ch] + 1
        seen[ch] = right
        max_len = max(max_len, right - left + 1)

    return max_len

print(length_of_longest_substring("abcabcbb"))
```

```python
"""Merge Intervals Problem:
Given intervals like [[1,3],[2,6],[8,10]]
→ merge overlapping ones Output: [[1,6],[8,10]]"""
def merge(intervals):
    intervals.sort()
    merged = []

    for i in intervals:
        if not merged or merged[-1][1] < i[0]:
            merged.append(i)
        else:
            merged[-1][1] = max(merged[-1][1], i[1])

    return merged

print(merge([[1,3],[2,6],[8,10]]))
```

```python
"""Product of Array Except Self
Problem: 
Return array where each element
= product of all elements except itself Constraint:
No division O(n) time"""
def product_except_self(nums):
    n = len(nums)
    res = [1]*n

    left = 1
    for i in range(n):
        res[i] = left
        left *= nums[i]

    right = 1
    for i in range(n-1, -1, -1):
        res[i] *= right
        right *= nums[i]

    return res

print(product_except_self([1,2,3,4]))
```

```python
"""Valid Parentheses Checker 
Problem: 
Check if string like "({[]})" is valid Must use stack logic"""
def is_valid(s):
    stack = []
    pairs = {')':'(', ']':'[', '}':'{'}

    for ch in s:
        if ch in pairs:
            if not stack or stack.pop() != pairs[ch]:
                return False
        else:
            stack.append(ch)

    return not stack

print(is_valid("({[]})"))
```

```python
"""Kth Largest Element Problem: 
Find kth largest element in array Constraint:
Don’t sort completely (optimize)"""
import heapq

def kth_largest(nums, k):
    return heapq.nlargest(k, nums)[-1]

print(kth_largest([3,2,1,5,6,4], 2))
```

```python
"""Search in Rotated Sorted Array Problem:
Array is sorted but rotated 
→ find target index
Example: [4,5,6,7,0,1,2], target=0
→ Output: 4"""
def search(nums, target):
    l, r = 0, len(nums)-1

    while l <= r:
        m = (l+r)//2

        if nums[m] == target:
            return m

        if nums[l] <= nums[m]:
            if nums[l] <= target < nums[m]:
                r = m-1
            else:
                l = m+1
        else:
            if nums[m] < target <= nums[r]:
                l = m+1
            else:
                r = m-1

    return -1

print(search([4,5,6,7,0,1,2], 0))
```

```python
"""Group Anagrams 
Problem: Group words that are anagrams 
Input: ["eat","tea","tan","ate","nat","bat"] 
Output: [["eat","tea","ate"],["tan","nat"],["bat"]]"""
from collections import defaultdict

def group_anagrams(words):
    d = defaultdict(list)

    for w in words:
        d[''.join(sorted(w))].append(w)

    return list(d.values())

print(group_anagrams(["eat","tea","tan","ate","nat","bat"]))
```

```python
"""Sliding Window Maximum (Hard) 
Problem: Find max in every window of size k
Example: [1,3,-1,-3,5,3,6,7], 
k=3 Output: [3,3,5,5,6,7]"""
from collections import deque

def max_sliding_window(nums, k):
    dq = deque()
    res = []

    for i in range(len(nums)):
        while dq and dq[0] <= i-k:
            dq.popleft()

        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()

        dq.append(i)

        if i >= k-1:
            res.append(nums[dq[0]])

    return res

print(max_sliding_window([1,3,-1,-3,5,3,6,7], 3))
```

```python
"""Minimum Platform Problem
Problem: Given arrival & departure times
→ find minimum platforms needed Classic greedy logic"""
def min_platforms(arr, dep):
    arr.sort()
    dep.sort()

    i = j = 0
    plat = maxp = 0

    while i < len(arr):
        if arr[i] <= dep[j]:
            plat += 1
            i += 1
        else:
            plat -= 1
            j += 1

        maxp = max(maxp, plat)

    return maxp

print(min_platforms([900,940,950,1100,1500,1800],
                    [910,1200,1120,1130,1900,2000]))
```

```python
"""Detect Cycle in List 
Problem: Check if cycle exists (Floyd’s Cycle Detection logic)"""
def has_cycle(head):
    slow = fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            return True

    return False
print(has_cycle)
```

```python
"""Top K Frequent Elements
Problem: Return k most frequent elements Optimize using hashmap + heap"""
from collections import Counter
import heapq

def top_k(nums, k):
    count = Counter(nums)
    return [i for i,_ in heapq.nlargest(k, count.items(), key=lambda x:x[1])]

print(top_k([1,1,1,2,2,3], 2))
```

```python
"""13. Word Break Problem (Dynamic Programming)
Problem:
Check if string can be segmented into dictionary words
Example:
"saptaparnichatterjee" → valid words list"""

def word_break(s, wordDict):
    word = set(wordDict)
    dp = [False]*(len(s)+1)
    dp[0] = True

    for i in range(len(s)):
        for j in range(i+1, len(s)+1):
            if dp[i] and s[i:j] in word:
                dp[j] = True

    return dp[-1]

print(word_break("leetcode", ["leet","code"]))
```

```python
"""14. Longest Palindromic Substring
Problem:
Find longest palindrome inside string
Example: "babad" → "bab" or "aba""""

def longest_palindrome(s):
    def expand(l,r):
        while l>=0 and r<len(s) and s[l]==s[r]:
            l-=1
            r+=1
        return s[l+1:r]

    res = ""

    for i in range(len(s)):
        res = max(res, expand(i,i), expand(i,i+1), key=len)

    return res

print(longest_palindrome("babad"))
```

```python
"""15. Trapping Rain Water 
Problem:
Given heights array → calculate trapped water
Example:
[0,1,0,2,1,0,1,3,2,1,2,1] → Output: 6"""
def trap(height):
    l, r = 0, len(height)-1
    left_max = right_max = 0
    water = 0

    while l < r:
        if height[l] < height[r]:
            if height[l] >= left_max:
                left_max = height[l]
            else:
                water += left_max - height[l]
            l += 1
        else:
            if height[r] >= right_max:
                right_max = height[r]
            else:
                water += right_max - height[r]
            r -= 1

    return water

print(trap([0,1,0,2,1,0,1,3,2,1,2,1]))
```
