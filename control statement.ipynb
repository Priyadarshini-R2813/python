```python
# sum of first n natural numbers
n=int(input("enter first number"))
sum=0
i=1
while i <=n:
    sum=sum+i
    i=i+1
    print("sum ",sum)
```

```python
#reverse a number
num=int(input("enter number"))
rev=0
while num>0:
    digit=num%10
    rev=rev*10+digit
    num=num//10
    print("reversed number is",rev)

```

```python
#count digit in number
num=int(input("enter a number:"))
count=0
if num==0:
    count=1
else:
    while num!=0:
         num=num//10
         count+=1
print("number of digits",count)
```

```python
#search for an element in a list and stop when found
list=[10,20,30,40,50]
key=int(input("enter element to search"))
for i in list:
    if i==key:
        print("element found:",i)
        break
    else:
        print("element not found")
```
