```python
''''Create a list with different data types:
integer, float, string, and boolean.
Print the list and its type. 
Print the type of each element. 
Example expected: my_list = [10, 3.5, "hello", True]'''

my_list = [10, 3.5, "hello", True]

print("List:", my_list)
print("Type of list:", type(my_list))

for item in my_list:
    print(item, "->", type(item))
```

```python
'''Create two lists: a = [1, 2, 3], b = [4, 5, 6] Perform: Concatenation Repetition (repeat list a 3 times) Membership check (check if 2 is in list a) Length of list'''
a = [1, 2, 3]
b = [4, 5, 6]

# Concatenation
print("Concatenation:", a + b)

# Repetition
print("Repetition:", a * 3)

# Membership check
print("Is 2 in list a?", 2 in a)

# Length
print("Length of list a:", len(a))
```

```python
"""nums = [10, 5, 20, 8, 15]
Perform: Find maximum value Find minimum value Find sum of elements Sort the list
(ascending & descending) 
Find index of element 8"""
nums = [10, 5, 20, 8, 15]
print("Maximum:", max(nums))
print("Minimum:", min(nums))
print("Sum:", sum(nums))
print("Ascending:", sorted(nums))
print("Descending:", sorted(nums, reverse=True))
print("Index of 8:", nums.index(8))
```

```python
''''Create a list: [10, 20, 30] Perform: Add element 40 using append() Insert 15 at index 1 Remove element 20 Pop last element Reverse the list'''
my_list = [10, 20, 30]

# Append
my_list.append(40)
print("After append:", my_list)

# Insert
my_list.insert(1, 15)
print("After insert:", my_list)

# Remove
my_list.remove(20)
print("After remove:", my_list)

# Pop
my_list.pop()
print("After pop:", my_list)

# Reverse
my_list.reverse()
print("After reverse:", my_list)
```

```python
"""Problem Solving
Take a list of numbers from user input
Print:
Even numbers
Odd numbers
Largest and smallest number
Bonus:
Remove duplicates from the list python program"""
# Taking input
nums = list(map(int, input("Enter numbers separated by space: ").split()))

# Even numbers
even = [n for n in nums if n % 2 == 0]
print("Even numbers:", even)

# Odd numbers
odd = [n for n in nums if n % 2 != 0]
print("Odd numbers:", odd)

# Largest & Smallest
print("Largest:", max(nums))
print("Smallest:", min(nums))

# Remove duplicates
unique = list(set(nums))
print("Without duplicates:", unique)
```
